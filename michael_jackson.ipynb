# 🎤 Michael Jackson - Historia y Legado

## 📌 Introducción
Michael Joseph Jackson (1958–2009), conocido como el "Rey del Pop", fue uno de los artistas más influyentes en la historia de la música. Su impacto abarcó la música, el baile, la moda y la cultura global.

---

## 👶 Infancia y comienzos
- Nació el **29 de agosto de 1958** en Gary, Indiana (EE.UU.).
- Fue el **octavo de diez hermanos** en la familia Jackson.
- Comenzó su carrera a temprana edad como miembro de **The Jackson 5**.

### 🎶 The Jackson 5
El grupo, formado junto a sus hermanos, alcanzó la fama con éxitos como:
- *I Want You Back*
- *ABC*
- *I'll Be There*

---

## 🌟 Carrera como solista
Michael Jackson lanzó su carrera en solitario en los años 70, pero alcanzó su mayor éxito en los años 80.

### 🏆 Álbumes más importantes
- **Off the Wall (1979)**
- **Thriller (1982)** → El álbum más vendido de la historia
- **Bad (1987)**
- **Dangerous (1991)**

---

## 💃 Innovaciones y estilo
Michael Jackson revolucionó la música y el entretenimiento:

- Popularizó el **Moonwalk**
- Innovó en los **videoclips musicales**
- Introdujo coreografías complejas en conciertos
- Fusionó géneros como pop, rock, soul y funk

---

## 🎥 Impacto en los videoclips
Sus videos musicales fueron clave en la historia del entretenimiento:

- *Thriller* → considerado el mejor videoclip de todos los tiempos
- *Beat It*
- *Billie Jean*

👉 Ayudó a convertir MTV en una plataforma global.

---

## 🌍 Impacto cultural
- Rompió barreras raciales en la música
- Influyó en generaciones de artistas
- Cambió la forma de hacer espectáculos en vivo

---

## ❤️ Trabajo humanitario
Michael Jackson apoyó múltiples causas benéficas:
- Donaciones a hospitales y organizaciones
- Canción **"We Are the World"** (1985)
- Inclusión en el Guinness por su labor solidaria

---

## ⚠️ Controversias
A lo largo de su vida enfrentó controversias personales y legales, que generaron debate mediático.

---

## 🕊️ Fallecimiento
- Murió el **25 de junio de 2009** en Los Ángeles, California.
- Su fallecimiento impactó al mundo entero.

---

## 🏅 Legado
- Más de **400 millones de discos vendidos**
- Múltiples premios (Grammy, American Music Awards)
- Reconocido como el **artista más exitoso de todos los tiempos**

---

## 📊 Curiosidades
- *Thriller* sigue siendo el álbum más vendido del mundo
- Fue pionero en fusionar música y cine
- Su estilo sigue influyendo en la cultura pop

Michael Jackson no solo transformó la música, sino también el entretenimiento global. Su legado continúa vivo a través de su música, su arte y su influencia en generaciones futuras.
