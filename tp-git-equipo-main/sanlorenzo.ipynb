<h1>San Lorenzo de Almagro</h1>

<h3><u>Nuevo Gasometro</u></h3>

<img src="Platea_Este_Nuevo_Gasometro_1.jpg" width=500px>
<p><a href="https://maps.app.goo.gl/XjS34YCiZ6V4yxLH9">Ubicación</a></p>
<p>El <b>Nuevo Gasómetro</b>, oficialmente denominado <b>Estadio Pedro Bidegain</b>, es el estadio de fútbol del Club Atlético San Lorenzo de Almagro, situado en el barrio Bajo Flores de la Ciudad de Buenos Aires. Fue inaugurado el 16 de diciembre de 1993, marcando el regreso del club a tener una cancha propia tras la pérdida del histórico Viejo Gasómetro de Avenida La Plata. Con una capacidad aproximada de <b>48.000 espectadores</b>, se destaca por poseer el campo de juego más grande del fútbol argentino, con medidas de <b>110 x 70 metros</b>, lo que le otorga una impronta única y lo convierte en uno de los escenarios más imponentes del país.

El estadio no solo es un espacio deportivo, sino también un símbolo de identidad y pertenencia para los hinchas del Ciclón, quienes lo consideran su verdadero hogar. A lo largo de su historia, el Nuevo Gasómetro fue testigo de innumerables partidos memorables, tanto en el ámbito local como internacional. Allí San Lorenzo conquistó títulos nacionales como los campeonatos Clausura 1995, Clausura 2001, Clausura 2007 e Inicial 2013, que devolvieron protagonismo al club en distintas épocas.

En el plano internacional, el estadio se convirtió en escenario de gestas históricas: la <b>Copa Mercosur 2001</b>, la <b>Copa Sudamericana 2002</b> y, especialmente, la <b>Copa Libertadores 2014</b>, el máximo logro de la institución, obtenida en su propia cancha frente a Nacional de Paraguay, con el recordado penal de Néstor Ortigoza que selló la gloria eterna. Estos títulos consolidaron al Nuevo Gasómetro como un lugar de prestigio continental y un templo del fútbol sudamericano.

Más allá de lo deportivo, el estadio también ha sido sede de eventos culturales, sociales y religiosos, reforzando su rol como punto de encuentro para la comunidad azulgrana y para la ciudad en general. Su arquitectura, con tribunas amplias y una disposición que favorece la visibilidad desde todos los sectores, refleja la magnitud de un club que supo sobreponerse a la adversidad y recuperar su casa.

En definitiva, el <b>Nuevo Gasómetro</b> es mucho más que un estadio: es un emblema de lucha, pasión y gloria, un espacio donde se escribieron páginas inolvidables de la historia de San Lorenzo y donde cada partido revive el sentimiento de pertenencia de miles de hinchas que lo consideran el corazón de su identidad futbolera.</p>

<h3><u>Trofeos</u></h3>

<img src="libertadores.jpg" width=500px>
<p><b>San Lorenzo de Almagro</b> vivió en el <b>Nuevo Gasómetro</b> algunos de los capítulos más gloriosos de su historia futbolística. En este estadio, inaugurado en 1993 y convertido rápidamente en un símbolo de identidad para la hinchada azulgrana, el club conquistó <b>cuatro campeonatos locales</b> que marcaron distintas épocas: el <b>Clausura 1995</b>, que rompió una sequía de seis años sin títulos; el <b>Clausura 2001</b>, bajo la conducción de Manuel Pellegrini, que devolvió protagonismo y abrió las puertas a la gloria internacional; el <b>Clausura 2007</b>, con Ramón Díaz como director técnico y un equipo que supo imponerse con carácter; y el <b>Inicial 2013</b>, que reafirmó la vigencia del club en el plano doméstico.

Pero la grandeza del Nuevo Gasómetro no se limita al ámbito nacional. Allí también se escribieron páginas doradas en el plano continental, con <b>tres títulos internacionales</b> que elevaron a San Lorenzo a la categoría de potencia sudamericana. En 2001, el club obtuvo la <b>Copa Mercosur</b>, su primer trofeo internacional, tras una final épica frente a Flamengo definida por penales. Al año siguiente, en 2002, llegó la <b>Copa Sudamericana</b>, conquistada frente a Atlético Nacional de Medellín, consolidando el prestigio del Ciclón en América.

El momento más emblemático, sin embargo, se produjo en 2014, cuando San Lorenzo alcanzó la cima del continente al ganar la <b>Copa Libertadores</b>, el máximo logro de su historia. La final frente a Nacional de Paraguay, disputada en el Nuevo Gasómetro, quedó grabada en la memoria colectiva: el penal ejecutado por Néstor Ortigoza selló la victoria y desató una celebración que trascendió lo deportivo, convirtiéndose en un hito cultural y emocional para toda la comunidad azulgrana.

Cada uno de estos títulos no solo representa un trofeo, sino también capítulos de lucha, pasión y orgullo que fortalecieron el vínculo entre el club y su gente. El Nuevo Gasómetro se consolidó así como un verdadero templo del fútbol argentino y sudamericano, un escenario donde se mezclan la historia, la emoción y la gloria eterna de San Lorenzo.

<h3><u>Jugadores Historicos</u></h3>

<p>El Club Atlético <b>San Lorenzo de Almagro</b> posee una rica historia marcada por grandes ídolos que dejaron huella en distintas épocas y que, con su talento y personalidad, forjaron la identidad azulgrana. Entre ellos se destaca <b>José Francisco Sanfilippo</b>, conocido como “El Nene”, máximo goleador histórico del club con más de 200 tantos, un delantero temido por todas las defensas y protagonista de clásicos inolvidables frente a River, Boca y Huracán. Su capacidad goleadora lo convirtió en símbolo de los años 50 y 60, y en uno de los referentes eternos de la institución.

Otro nombre fundamental es <b>Roberto Telch</b>, apodado “La Oveja”, un mediocampista incansable que brilló en los años 60 y 70. Telch fue parte de los legendarios “Carasucias” y de los “Matadores” campeones de 1968, aportando equilibrio, garra y calidad técnica. Su presencia en la Selección Argentina, incluso en el Mundial de 1974, refleja la magnitud de su carrera y su importancia en el fútbol nacional.

En el plano de la regularidad y la fidelidad, sobresale <b>Sergio Villar</b>, defensor uruguayo que ostenta el récord absoluto de partidos disputados con la camiseta de San Lorenzo: <b>446 encuentros oficiales</b>. Su constancia y compromiso lo convirtieron en un verdadero emblema de la institución, respetado por compañeros y rivales.

La historia azulgrana también se nutre de figuras de épocas más lejanas, como <b>Ángel Zubieta</b>, quien llegó desde España en 1939 y rápidamente se transformó en capitán y referente del equipo campeón de 1946. Su nobleza, juego limpio y liderazgo lo convirtieron en un símbolo de aquellos años, recordado como uno de los grandes caballeros del fútbol argentino.

En tiempos modernos, el ídolo indiscutible es <b>Leandro “Pipi” Romagnoli</b>, surgido de las divisiones inferiores y protagonista de la etapa más gloriosa del club en el siglo XXI. Capitán en la histórica <b>Copa Libertadores 2014</b>, Romagnoli levantó el trofeo más importante de la historia azulgrana y se consolidó como el jugador más ganador de la era profesional, con seis títulos en su haber. Su talento, fidelidad y amor por la camiseta lo convirtieron en un ícono contemporáneo y en el puente entre la tradición y la modernidad del club.

Cada uno de estos ídolos representa un estilo y una época: Sanfilippo como el goleador letal, Telch como el motor del mediocampo, Villar como el símbolo de la constancia, Zubieta como el capitán noble de los 40 y Romagnoli como el líder de la gloria internacional. Juntos conforman un mosaico de figuras que reflejan la evolución de San Lorenzo y que mantienen viva la pasión de generaciones enteras de hinchas.

En definitiva, la larga lista de ídolos azulgranas es el testimonio de un club que supo construir su identidad a través de hombres que dejaron todo en la cancha y que, con sus gestas, se ganaron un lugar eterno en la memoria colectiva de la hinchada.</p>
