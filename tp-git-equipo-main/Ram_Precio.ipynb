## La primera subida en el precio de la ram (2017-2018):

Esta primera subida fue debida a que los grandes proveedores de servicios de computacion en la nube 
(por ejemplo: Amazon Web Services, Microsoft Azure, Google Cloud) construyeron grandes centros de datos y sobrepasaron la oferta disponible en esa epoca.
Esto represento un aumento del 62% en 2017 y de 27% en 2018.

## Caída de precios (2020–2023):

durante estos años el precio de la memoria ram empezo a estabilizarse y a bajar, esto es debido a la baja en la demanda y el aumento de produccion, logrando que esta llegue a un precio accesible.

## Gran subida reciente (2024–2026)

En la actualidad el precio de la ram se disparo debido a el entrenamiento al auge de la inteligencia artificial. Esto se debe a que
entrenar modelos de inteligencia artificial requiere de enormes cantidades de memoria, entonces las grandes empresas de IA compran grandes cantidades
de chips de memoria directamente de los productores (ya que estas pagan mas) haciendo que la mayoria de la produccion se centre en chips para centros de datos que son vendidos de manera directa a estas empresas, dejando de lado la produccion para el consumidor. Esto genero que a dia de hoy dos modulos de ram ddr5 (16 GB) lleguen a estar 400 dolares
