# La Guerra del Emú 

## ¿De que trato? 

la guerra del Emú ocurrio después de la Gran Depresión en el año 1932, fue un suceso surrealista en Australia que duro un mes, donde el ejército, armado con ametralladoras, fracasó en su intento de controlar una plaga de 20,000 emús que destruían cultivos en Australia Occidental

## Contexto
Despues de la primera guerra mundial, muchos exsoldados recibieron tierras para cultivar en Australia Occidental. El problema fue que una migracion de emúes (aves grandes parecidas al avestruz) estaban por la zona y entraban en los campos, destruyendo cultivos y cercos.
Los granjeros pidieron ayuda al gobierno, y el gobierno decidio tratar de solucionar el problema, hubieron un total de 2 intentos.

![alt text](<muchos emus-1.jfif>)

## Primer intento
Durante la Guerra del Emú, el mayor Gwynydd Purves Wynne-Aubrey Meredith, junto al sargento S. McMurray y el artillero J. O'Halloran, intentó emboscar y eliminar grupos de emús. Sin embargo, las aves se dispersaban y eran difíciles de alcanzar, y aunque lograron abatir algunas, las ametralladoras fallaban o resultaban poco efectivas, incluso al intentar perseguirlos en camión. Tras varios días y miles de disparos, los resultados fueron limitados. Finalmente, tras críticas y debate político impulsado por figuras como George Pearce, la operación fue cancelada, destacándose la resistencia y movilidad de los emús.

## Segundo intento
Tras la retirada inicial en la Guerra del Emú, los emús retomaron los ataques a los cultivos, lo que llevó a los agricultores a pedir nuevamente ayuda debido a la sequía y el calor. El primer ministro James Mitchell apoyó la solicitud, y el ministro de Defensa George Pearce autorizó reanudar la operación el 12 de noviembre, manteniendo al mayor Gwynydd Purves Wynne-Aubrey Meredith como comandante.

En esta segunda fase, los soldados lograron mejores resultados al principio, con decenas de emús abatidos en los primeros días. Aunque hubo jornadas menos efectivas, para diciembre el ritmo de bajas aumentó, alcanzando cerca de cien emús por semana. Finalmente, el 10 de diciembre, Meredith fue retirado y reportó unas 986 aves muertas con casi 10.000 disparos, además de miles de emús heridos que probablemente murieron después.

## Resultados
Tras la Guerra del Emú, los agricultores siguieron pidiendo ayuda militar en varias ocasiones, pero el gobierno la rechazó. En su lugar, se utilizó un sistema de recompensas que resultó más efectivo, permitiendo abatir decenas de miles de emús y reduciendo significativamente su población con el tiempo.
La noticia del conflicto también llegó al Reino Unido, donde generó críticas por parte de conservacionistas, quienes denunciaron la matanza como una exterminación masiva, una opinión reforzada por el ornitólogo Dominic Serventy.

Al final, el ejército se retiró sin lograr controlar a los emús en la Guerra del Emú, ya que las aves lograban escapar y resistir los ataques, por lo que se las considera “ganadoras” del conflicto. Debido a este fracaso, el gobierno dejó de enviar militares, y aunque los agricultores siguieron pidiendo ayuda, sus solicitudes fueron rechazadas. En su lugar, se implementó un sistema de recompensas que resultó más efectivo para reducir la población. Sin embargo, la matanza generó críticas internacionales, especialmente en el Reino Unido, donde conservacionistas y el ornitólogo Dominic Serventy la calificaron como una exterminación masiva.
