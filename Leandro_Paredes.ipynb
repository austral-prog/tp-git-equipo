### El pibe que heredó la 10
Leandro Daniel Paredes surgiió de la inagotable cantera de Boca Juniors. Su talento era tan evidente que hizo su debut profesional el 6 de noviembre de 2010, con apenas 16 años, bajo la dirección técnica de Claudio Borghi. Desde el primer momento en que pisó el césped de La Bombonera, los hinchas notaron algo distinto. Su visión de juego periférica, el control orientado y una pegada envidiable hicieron que rápidamente lo etiquetaran como el sucesor natural de Juan Román Riquelme. De hecho, el propio Román lo apadrinó en sus primeros años, y Leandro no se achicó cuando le tocó llevar la mítica camiseta número 10 de Boca antes de dar su primer gran salto a Europa.

### La transformación europea: De enganche a "Cinco" moderno

![Paredes en la Selección](https://platform.chiesaditotti.com/wp-content/uploads/sites/64/chorus/uploads/chorus_asset/file/25516509/2159988852.jpg?quality=90&strip=all&crop=0,4.0566666666667,100,44.355555555556)

El fútbol europeo exige una intensidad física y táctica diferente, y Paredes supo reinventarse con una inteligencia asombrosa. En su paso por la Serie A con la Roma y el Empoli, y más tarde en el Zenit de Rusia, dejó de ser el clásico enlace libre de responsabilidades defensivas para convertirse en un volante central moderno y posicional. 

Su consolidación mundial llegó en el Paris Saint-Germain (PSG) y la Juventus. En estos clubes de élite, perfeccionó su rol de "5": se volvió un experto en el posicionamiento táctico, un recuperador feroz cuando el partido exige meter la pierna, y el primer pase limpio para la salida del equipo. Esta metamorfosis fue la clave que le abrió las puertas a la gloria internacional.

### El corazón del mediocampo Campeón del Mundo
La verdadera leyenda de Leandro Paredes se escribió con la camiseta de la Selección Argentina. Fue el jugador fetiche del inicio del ciclo de Lionel Scaloni, formando un triángulo en la mitad de la cancha que le devolvió la identidad al equipo. 

Fue fundamental en la conquista de la **Copa América 2021** en el Maracaná, rompiendo el maleficio de 28 años sin títulos. Al año siguiente, en el **Mundial de Qatar 2022**, Paredes demostró que tiene sangre fría en los momentos donde a otros les queman las piernas. Sus ejecuciones en las definiciones por penales frente a Países Bajos (cuartos de final) y Francia (en la final) fueron remates potentes y esquineros, mostrando un temple absoluto para ayudar a bordar la tercera estrella en el escudo argentino.

### El regreso a casa: Vigencia y jerarquía física

![Paredes regreso a Boca](https://lanumero12.com.ar/storage/2025/10/Leandro-Paredes-Boca-201025.jpg.webp)

A diferencia de muchos futbolistas que deciden volver al fútbol sudamericano en el ocaso de sus carreras o cuando el físico ya no les responde, Leandro Paredes pegó la vuelta a Boca Juniors cumpliendo una promesa que siempre le hizo al hincha: **volver en plenitud**. 

Su retorno a La Ribera se dio en un estado físico impecable. Lejos de venir a buscar un retiro cómodo, Paredes volvió con el ritmo de la élite europea, soportando sin problemas el roce, la fricción y el calendario apretado del fútbol argentino. Rápidamente se adueñó de la mitad de la cancha, demostrando que su calidad sigue intacta y que el tanque de nafta está lleno. Verlo barrer, recuperar y meter pases filtrados de 40 metros en La Bombonera confirmó lo que todos esperaban: volvió un jugador de jerarquía mundial, entero físicamente, con hambre de ganar la Copa Libertadores y de seguir haciendo historia en el club de sus amores.
