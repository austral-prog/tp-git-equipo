# Galgos

![Galgo](Assets/galgo.jpg "Galgo")

## Intro
Los galgos son perros ligeros y rápidos. Se suelen entrenar para competir. Estas competencias son controversiales debido al maltrado de galgos. Ultimamente se estan prohibiendo estas carreras y con el tiempo el galgo aparece más en las casas como mascota.

![Galgo 2](Assets/galgo2.jpg "Galgo")

## Característica
Son altos y/o largos. Su pelaje es corto, liso y muy fino. Otra raza, el borzoi, es una version peluda del galgo.
