# Tutorial basico de impresion 3d

## Primeros pasos

Primero revisar y seguir las instrucciones de instalacion que vienen con la impresora.
Introducir el filamento en la impresora, se recomienda empezar con PLA ya que es el mas facil de utilizar.
Elegir uno de los modelos que vienen pre-programdos en la impresora para probar que este todo en orden.
Si hay algun error revise que todo este instalado correctamente, si el error persiste el problema podria ser el filamento, si puede pruebe con un rollo distinto, si no tiene otro rollo tendra que secar el filamento, si su impresora vino con secadora de filamento usela sino, avance mas adelante que se hablara de un metodo de como secarlo.
Si no hay ningun error ya puede empezar con el siguiente paso.

## Problemas comunes

Hay tres lugares de donde pueden aparecer los problemas que son:
* El equipo
* El filamento
* La configuracion
La habilidad mas importante es poder reconocer de donde viene el problema, los casos mas probables son los dos primeros(el equipo y el filamento) aunque dependiendo de la impresora que tengas esto podria cambiar, para poder determinar de donde viene el error aca hay un par de ejemplos y su causa.
